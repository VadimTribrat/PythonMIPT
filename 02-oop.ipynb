Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
from math import atan, cos, sin
import numpy as np
import math


class ComplexNumber:
    def __init__(self, i=0, j=0, polar=False):
        if polar:
            self._i = i * cos(j)
            self._j = i * sin(j)
        else:
            self._i = i
            self._j = j
    
    def __add__(self, other):
        return ComplexNumber(self._i + other._i, self._j + other._j)

    def __mul__(self, other):
        temp_i = self._i * other._i - self._j * other._j
        temp_j = self._i * other._j + self._j * other._i
        return ComplexNumber(temp_i, temp_j)

    def __sub__(self, other):
        return ComplexNumber(self._i - other._i, self._j - other._j)

    def __truediv__(self, other):
        norm = other._i ** 2 + other._j ** 2
        if not norm:
            raise ZeroDivisionError("ZeroDivisionError")
        temp_i = self._i * other._i + self._j * other._j
        temp_j = other._i * self._j - self._i * other._j
        return ComplexNumber(temp_i / norm, temp_j / norm) 

    def __eq__(self, other):
        return math.isclose(self._i, other._i) and math.isclose(self._j, other._j)

    def __repr__(self):
        return "{0} {1}i".format(self._i, self._j)

    def get_algebraic_form(self):
        return (self._i, self._j)

    def get_polar_form(self):
        norm = self._i ** 2 + self._j ** 2
        if not norm:
            raise ZeroDivisionError()
        return (norm ** 0.5, atan(self._j / self._i))
````

Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
class Vector:
    def __init__(self, vector_values_list):
        if isinstance(vector_values_list, list):
            self._values = vector_values_list
        else:
            raise ValueError("ValueError")

    def __add__(self, other):
        if len(self._values) != len(other._values):
            raise ValueError("ValueError")
        return Vector([i + j for i, j in zip(self._values, other._values)])

    def __sub__(self, other):
        if len(self._values) != len(other._values):
            raise ValueError("ValueError")
        return Vector([i - j for i, j in zip(self._values, other._values)])

    @staticmethod
    def scalar_product(arg1, arg2):
        if len(arg1._values) != len(arg2._values):
            raise ValueError("ValueError")
        return sum([i * j for i, j in zip(arg1._values, arg2._values)])

    @staticmethod
    def cosine(self, other):
        scalar_prod = Vector.scalar_product(self, other)
        arg1 = self.euclid_norm()
        arg2 = other.euclid_norm()
        if not (arg1 or arg2):
            raise ZeroDivisionError("ZeroDivisionError")
        return scalar_prod / (arg1 * arg2)

    def euclid_norm(self):
        return Vector.scalar_product(self, self) ** 0.5

    def __repr__(self):
        return str(self._values)
````

````python
test1 = np.random.uniform(-10, 10, size=5)
test2 = np.random.uniform(-10, 10, size=5)
test3 = np.random.uniform(-10, 10, size=7)
vec1 = Vector(test1.tolist())
vec2 = Vector(test2.tolist())
vec3 = Vector(test3.tolist())
print(vec1, vec2, vec3, sep='\n')

assert np.all(np.isclose((vec1 + vec2)._values, test1 + test2))
assert np.all(np.isclose((vec1 - vec2)._values, test1 - test2))
assert np.isclose(Vector.scalar_product(vec1, vec2), test1 @ test2)
assert vec1.euclid_norm() == np.linalg.norm(test1)
assert Vector.cosine(vec1, vec2) == test1 @ test2 / (np.linalg.norm(test1) * np.linalg.norm(test2))
try:
    Vector(set([1, 2, 3]))
except ValueError as e:
    print(e)
try:
    vec1 + vec3
except ValueError as e:
    print(e)
try:
    Vector.cosine(vec1, Vector([0 for i in range(5)]))
except ZeroDivisionError as e:
    print(e)
````

__Задание 4 (2 балл):__ Опишите декоратор, который принимает на вход функцию и при каждом её вызове печатает строку "This function was called N times", где N - число раз, которое это функция была вызвана на текущий момент (пока функция существует как объект, это число, очевидно, может только неубывать).

````python
from functools import wraps


def calls_counter(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        wrapper.counter += 1
        print("Function {0} was called {1} times".format(wrapper.__name__, wrapper.counter))
        return func(*args, **kwargs)
    wrapper.counter = 0
    return wrapper

@calls_counter
def test():
    pass

@calls_counter
def test1():
    pass

for i in range(7):
    test()

for i in range(5):
    test1()
````

__Задание 5 (3 балла):__ Опишите декоратор класса, который принимает на вход другой класс и снабжает декорируемый класс всеми атрибутами входного класса, названия которых НЕ начинаются с "\_". В случае конфликтов имён импортируемый атрибут должен получить имя с суффиксом "\_new".

````python
def copy_class_attrs(cls):
    def decorator(cls_1):
        set_ = set([val for val in dir(cls) if not val.startswith('_')])
        set_1 = set(dir(cls_1))
        intersect = set.intersection(set_, set_1)
        for val in intersect:
            setattr(cls_1, "_new" + val, getattr(cls, val))
        new_attr = set_.difference(intersect)
        for val in new_attr:
            setattr(cls_1, val, getattr(cls, val))
        return cls_1
    return decorator
````

````python
class A:
    def __init__(self):
        self.a = 6
    
    @staticmethod
    def foo():
        print("I'm from A")
    
    def foo1(self):
        print("I'm foo1 from A")

@copy_class_attrs(A)
class B:
    def __init__(self):
        self.b = 6

    @staticmethod
    def foo():
        print("I'm from B")
        
    def foo2(self):
        print("I'm foo2 from B")

dir(B)[-4:], dir(A)[-2:]
````

````python
import pickle
from copy import deepcopy


class SaveAsText:
    def save_as_text(self, filename):
        with open(filename, 'w') as f:
            f.write(str(self))

class SaveAsBin:
    def save_as_bin(self, filename):
        with open(filename, 'wb') as f:
            pickle.dump(self, f)

class LoadAsText:
    @staticmethod
    def load_as_text(filename):
        with open(filename, 'r') as f:
            new_mat = []
            for line in f:
                new_mat.append(line.split())
            return Matrix(new_mat)

class LoadAsBin:
    @staticmethod
    def load_as_bin(filename):
        with open(filename, 'rb') as f:
            return pickle.load(f)

class DeepCopy:
    def deep_copy(self):
        return deepcopy(self)

class Matrix(DeepCopy, SaveAsText, LoadAsText, SaveAsBin, LoadAsBin):
    def __init__(self, list_, b=0, c=0):
        if isinstance(list_, list):
            if isinstance(list_[0], list):
                self.shape = len(list_), len(list_[0])
                self.values = []
                for row in list_:
                    self.values.extend(row)
            else:
                self.shape = 1, len(list_)
                self.values = list_ 
        if isinstance(list_, float) or isinstance(list_, int):
            if list_ <= 0 or b <= 0:
                raise ValueError("ShapeError")
            self.shape = (list_, b)
            self.values = [c for _ in range(b*list_)]

    def __getitem__(self, ind):
        ind1, ind2 = ind
        if not (0 <= ind1 < self.shape[0] and 0 <= ind2 < self.shape[1]):
            raise IndexError("IndexError")
        return self.values[ind1 * self.shape[1] + ind2]

    def __add__(self, other):
        if self.shape != other.shape:
            raise ValueError("ShapeError")
        new_val = []
        for i in range(self.shape[0]):
            temp = []
            for j in range(self.shape[1]):
                temp.append(self[i, j] + other[i, j])
            new_val.append(temp)
        return Matrix(new_val)

    def __sub__(self, other):
        if self.shape != other.shape:
            raise ValueError("ShapeError")
        new_val = []
        for i in range(self.shape[0]):
            temp = []
            for j in range(self.shape[1]):
                temp.append(self[i, j] - other[i, j])
            new_val.append(temp)
        return Matrix(new_val)

    def __mul__(self, other):
        if self.shape != other.shape:
            raise ValueError("ShapeError")
        new_val = []
        for i in range(self.shape[0]):
            temp = []
            for j in range(self.shape[1]):
                temp.append(self[i, j] * other[i, j])
            new_val.append(temp)
        return Matrix(new_val)                

    def __eq__(self, other):
        if self.shape != other.shape:
            raise ValueError("ShapeError")
        for i in range(self.shape[0]):
            for j in range(self.shape[1]):     
                if self[i, j] != other[i, j]:
                    return False
        return True

    def __str__(self):
        return self.__repr__()

    def __repr__(self):
        string = ""
        for row in range(self.shape[0]):
            for val in range(self.shape[1]):
                string += " " + str(self.values[row * self.shape[1] + val])
            string += "\n"
        return string

    def dot(self, other):
        if self.shape[1] != other.shape[0]:
            raise ValueError("ShapeError")
        new_val = []
        for i in range(self.shape[0]):
            temp = []
            for j in range(other.shape[1]):
                sum_ = 0
                for k in range(self.shape[1]):
                    sum_ += self[i, k] * other[k, j]
                temp.append(sum_)
            new_val.append(temp)
        return Matrix(new_val)

    def trace(self):
        if self.shape[0] != self.shape[1]:
            return None
        k = min(self.shape[0], self.shape[1])
        sum_ = 0
        for i in range(k):
            sum_ += self[i, i]
        return sum_

    def transpose(self):
        new_val = []
        for i in range(self.shape[1]):
            new_val.append(self.values[i : : self.shape[1]])
        return Matrix(new_val)

    def det(self):
        if self.shape[0] != self.shape[1]:
            print("ShapeError")
            return None
        new_val = []
        for i in range(self.shape[0]):
            temp = []
            for j in range(self.shape[1]):
                temp.append(self[i, j])
            new_val.append(temp)
        for j in range(self.shape[1] - 1):
            ind = -1
            i = j
            while (ind==-1) and (i<self.shape[0]):
                if new_val[i][j] != 0:
                    ind = i
                    break
                i += 1     
            if ind == -1:
                return 0
            new_val[j], new_val[ind] = new_val[ind], new_val[j]
            for k in range(j + 1, self.shape[0]):
                Matrix.__sub_row(new_val, k, j, new_val[k][j]/new_val[j][j])
        res = 1
        for k in range(len(new_val)):
            res *= new_val[k][k]
        return res

    @staticmethod
    def __sub_row(mat, k1, k2, coef):
        for i in range(len(mat[0])):
            mat[k1][i] = mat[k1][i] - mat[k2][i] * coef

    def iter_by_row(self):
        new_val = []
        for i in range(self.shape[0]):
            new_val.append(self.values[i * self.shape[1] : (i + 1) * self.shape[1]])
        return new_val

    def iter_by_col(self):
        new_val = []
        for i in range(self.shape[1]):
            new_val.append(self.values[i : : self.shape[1]])
        return new_val

# def pr(mat):
#     for row in mat:
#         for val in row:
#             print(val, end=' ')
#         print()
#     print()
````

Каждый класс должен характеризоваться методом вычисления стоимости чашки `calculate_cost`. Пользователь должен иметь возможность комбинировать любое число добавок с выбранным кофе и получить на выходе общую стоимость:

```
Cream(Sugar(Latte())).calculate_cost()
```

Первым элементом чашки всегда должен быть сам кофе, а не добавка, в противном случае при попытке создания чашки должно выбрасываться исключение:

```
Cream(Latte(Sugar())).calculate_cost() -> exception
```

Кофе может встречаться в чашке только один раз, в противном случае при попытке создания чашки должно выбрасываться исключение:

```
Cappuccino(Sugar(Latte())).calculate_cost() -> exception
```

Добавки могут включаться в чашку в любом количестве и порядке.
Добавление новых типов кофе и добавок не должно требовать изменения существующего кода.

````python
from abc import ABC, abstractmethod


class Interface(ABC):
    @abstractmethod
    def calculate_cost(self):
        pass

class Latte(Interface):
    def __init__(self, val=None):
        self.price = 10
        if val is not None:
            self.val = val
            setattr(self.val, 'coffee', True)

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
            raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        return res

class Cappuccino(Interface):
    def __init__(self, val=None):
        self.price = 15
        if val is not None:
            self.val = val
            setattr(self.val, 'coffee', True)

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
            raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        return res

class Americano(Interface):
    def __init__(self, val=None):
        self.price = 20
        if val is not None:
            self.val = val
            setattr(self.val, 'coffee', True)

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
            raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        return res

class Sugar(Interface):
    def __init__(self, val=None):
        self.price = 1
        if val is not None:
            self.val = val

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
                raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        else:
            raise ValueError("ValueError")
        return res

class Cream(Interface):
    def __init__(self, val=None):
        self.price = 1
        if val is not None:
            self.val = val

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
                raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        else:
            raise ValueError("ValueError")
        return res

class Cardamom(Interface):
    def __init__(self, val=None):
        self.price = 1
        if val is not None:
            self.val = val

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
                raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        else:
            raise ValueError("ValueError")
        return res

class Foam(Interface):
    def __init__(self, val=None):
        self.price = 1
        if val is not None:
            self.val = val

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
                raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        else:
            raise ValueError("ValueError")
        return res

class Syrup(Interface):
    def __init__(self, val=None):
        self.price = 1
        if val is not None:
            self.val = val

    def calculate_cost(self):
        res = self.price
        if hasattr(self, 'coffee'):
                raise ValueError("ValueError")
        if hasattr(self, 'val'):
            res += self.val.calculate_cost()
        else:
            raise ValueError("ValueError")
        return res
````
